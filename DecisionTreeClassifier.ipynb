## Exploring Decision Tree Algorithm
#### For the given ‘Iris’ dataset, create the Decision Tree classifier and visualize it graphically. The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly

### Importing Required libraries

```python
import pandas as pd
import seaborn as sns
```

### Reading the dataset

```python
df = pd.read_csv('iris.csv')
df = df.drop(['Id'], axis = 1)
df.head()
```

### Checking for any null values

```python
df.isna().sum()
```

### Preparing the data

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,4].values
```

### Visualising the Dataset 

```python
sns.pairplot(df,hue="Species")
```

### Dividing the data into training and testing sets

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=1)
```

### Creating Decision Tree Clasifier

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=4)
model.fit(x_train, y_train) 
```

### Checking the accuracy of the classifier

```python
y_pred=model.predict(x_test)
from sklearn.metrics import accuracy_score
print(f"Accuracy of the classifier = {accuracy_score(y_pred,y_test)*100:.2f}%")
```

### Predicting the Species for new values

```python
y_pred=model.predict([[5.8,2.2,4.2,1.3]])
print(f"The species is : {y_pred[0]}")
```
